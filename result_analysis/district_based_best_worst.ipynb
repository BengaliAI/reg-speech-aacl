```python
import pandas as pd
import os
import matplotlib.pyplot as  plt
import numpy as np
pd.set_option('display.max_colwidth', 0)
```

```python
filenames = os.listdir('benchmarks/Competition Models/')
filenames = [f for f in filenames if '.csv' in f]
```

```python
filenames
```

```python
df_all = []
for f in filenames:
    temp = pd.read_csv('benchmarks/Competition Models/{}'.format(f)) 
    temp['model'] = f
    df_all.append(temp)
df_all = pd.concat(df_all)
```

```python
df_all
```

```python
worst_performing = None
for model in filenames:
    # print('=========================')
    # print(f)
    df = df_all.groupby('model').get_group(model)
    thresh = np.percentile(df['wer'],90.0)
    if worst_performing is None:
        worst_performing = set(df[df['wer']>thresh]['id'].values)
    else:
        set_1 = set(df[df['wer']>thresh]['id'].values)
        # print(worst_performing)
        # print(set_1)
        worst_performing = worst_performing.intersection(set_1)
        # print(worst_performing)
print(len(worst_performing))
print(worst_performing)
```

```python
df_all[[item in worst_performing for item in df_all['id']]]
```

```python
best_performing = None
for model in filenames:
    # print('=========================')
    # print(f)
    df = df_all.groupby('model').get_group(model)
    thresh = np.percentile(df['wer'],20.0)
    # print(thresh)
    if best_performing is None:
        best_performing = set(df[df['wer']<thresh]['id'].values)
    else:
        set_1 = set(df[df['wer']>thresh]['id'].values)
        # print(worst_performing)
        # print(set_1)
        best_performing = best_performing.intersection(set_1)
        # print(worst_performing)
print(len(best_performing))
print(best_performing)
```

```python
df_all[[item in best_performing for item in df_all['id']]]
```
