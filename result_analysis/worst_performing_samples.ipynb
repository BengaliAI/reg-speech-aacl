```python
import pandas as pd
import os
import matplotlib.pyplot as  plt
import numpy as np
```

```python
filenames = os.listdir('benchmarks/Competition Models')
```

```python
filenames
```

```python
df_all = []
for f in filenames:
    temp = pd.read_csv('benchmarks/Competition Models/{}'.format(f)) 
    temp['model'] = f
    df_all.append(temp)
df_all = pd.concat(df_all)
```

```python
df_all
```

```python
for model in filenames:
    print('=========================')
    print(f)
    df = df_all.groupby('model').get_group(f)
    print(df[['wer', 'cer']].describe())
    print(np.percentile(df['wer'],99.0))
    
```

```python
worst_performing = None
for model in filenames:
    # print('=========================')
    # print(f)
    df = df_all.groupby('model').get_group(f)
    thresh = np.percentile(df['wer'],99.0)
    if worst_performing is None:
        worst_performing = set(df[df['wer']>thresh]['id'].values)
    else:
        set_1 = set(df[df['wer']>thresh]['id'].values)
        # print(worst_performing)
        # print(set_1)
        worst_performing = worst_performing.intersection(set_1)
        # print(worst_performing)
print(len(worst_performing))
print(worst_performing)
```

```python
for id_ in worst_performing:
    df = df_all.groupby('id').get_group(id_)
    df['wer'].plot.bar()
    plt.xticks(range(len(df['model'])),df['model'].values)
    plt.title(id_)
    plt.show()
```
